```python
import nbformat
from IPython import get_ipython
from IPython.core.interactiveshell import InteractiveShell

def run_notebook(notebook_path):
    with open(notebook_path) as f:
        notebook_content = nbformat.read(f, as_version=4)
    
    # Create an interactive shell
    shell = InteractiveShell.instance()    
    for cell in notebook_content.cells:
        if cell.cell_type == 'code':
            code = cell.source
            shell.run_cell(code)

run_notebook('Q1.ipynb')
run_notebook('Q2.ipynb')
run_notebook('Q3.ipynb')
```

```python
def InvokeModel(plot):
    initial_portfolio_value = 10000  
    years = 10
    num_simulations = 1000

    # use pessimistic estimate of BTC as distribution is much more spread than FTSE
    R = r
    assets = ['FTSE', 'BTC']
    expected_returns = np.array([ftse_annual, btc_annual])  
    volatilities = np.array([ftse_vol_annual, btc_75]) 
    print(assets, expected_returns, volatilities, R)

    # Test out different iterations of portfolio weights
    n = 10
    final_p_5s = {}
    final_medians = {}
    final_p_95s = {}

    for i in range(n+1):
        weight_a = i/n
        weight_b = 1 - weight_a
        weights = np.array([weight_a, weight_b]) 

        print(weights)
        mode, mean, median, p_5, p_95 = MonteCarlo(initial_portfolio_value=initial_portfolio_value,
                                        years = years,
                                        num_simulations=num_simulations,
                                        expected_returns=expected_returns,
                                        volatilities=volatilities,
                                        weights=weights,
                                        R=R,
                                        plot=plot)
        final_p_5s[weight_a] = p_5
        final_medians[weight_a] = median
        final_p_95s[weight_a] = p_95
        print()

    plt.title('Simulated return for FTSE & BTC Portfolio with £10k initial investment after 10Y at 5th Percentile')
    PlotExpectedReturns(assets=assets, metrics=final_p_5s)
    plt.show()

# Show inner workings once
InvokeModel(plot=True)
```
