```python
import yfinance as yf
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
```

```python
# Fetch the ticker data for AstraZeneca
azn_lond = yf.Ticker('AZN.L')
azn_lond_hist = azn_lond.history(period="5y")
```

```python
azn_lond_hist = azn_lond_hist.reset_index()
azn_lond_hist['Date'] = pd.to_datetime(azn_lond_hist['Date']).dt.date
azn_lond_hist
```

```python
azn_lond_hist.describe()
```

```python
date = azn_lond_hist['Date']
close = azn_lond_hist['Close']

# Plot the data
plt.figure()
plt.plot(date, close, marker='.', linestyle='-')
plt.title('AstraZeneca (AZN) Closing Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price (GBX)')
plt.xticks(rotation=45)  # Rotate date labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of tick-labels
plt.show()
```
